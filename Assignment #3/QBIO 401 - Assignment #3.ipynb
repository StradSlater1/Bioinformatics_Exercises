# Problem 1 - Alighnment Agreement

```python
def read_multiz(filename):
    blocks = []
    multiz = open(filename, 'r').readlines()
    last_line = multiz[-1]
    i = 2
    while (i < len(multiz) ):
        block = []
        while(multiz[i][0] != '\n'):
            
            if((multiz[i][0] == 'i') or (multiz[i][0] == 'e') or (multiz[i][0] == 'a')):
                pass
            else:
                block.append(multiz[i].split()[-1])
            i += 1
        blocks.append(block)
        i+= 1
    return blocks
```

```python
def agree_with_human(species, human, k_values):
    for j in range(len(human)):
        amount = 0
        for i in range(len(species)):
            if species[i][j] == human[j]:
                amount += 1
        value = k_values.get(amount)
        value += 1
        k_values[amount] = value
        
    return k_values
```

### Test 

```python
species = ["ACGTTAT", "ACGTTTT", "AC-----"]
human = "ACGTTAC"
k_values = {}
for i in range(4):
    k_values[i] = 0
    
agree_with_human(species, human, k_values)

first_key = next(iter(k_values))
for key, value in k_values.items():
    print("k=", key, " # = ", value)
```

# Problem 2 - Reporting the Nucleotide Positions

```python
def report_nucleotides(species, human, k_values, addition):
    for j in range(len(human)):
        amount = 0
        for i in range(len(species)):
            if species[i][j] == human[j]:
                amount += 1
        k_values[amount].append(j+addition)
    return k_values
```

### Test 

```python
species = ["ACGTTAT", "ACGTTTT", "AC-----"]
human = "ACGTTAC"
k_values = {}
for i in range(4):
    k_values[i] = []
    
report_nucleotides(species, human, k_values,0)

first_key = next(iter(k_values))
for key, value in k_values.items():
    print("k=", key, " positions = ", value)
```

# Problem 3 - Counting Bases and Indels

```python
def count_base_indel(blocks):
    base_amount = 0
    indel_amount = 0
    for block in blocks:
            for i in range(len(block[0])):
                if(block[0][i] == '-'):
                    indel_amount += 1;
                else:
                    base_amount += 1;
    return [base_amount, indel_amount]
```
