## Problem 1 - Mapping Genome Builds to Species Names

```r
library(GenomeInfoDb)
library(yaml)
```

```r
map_genome_builds <- function(file) {
  data <- read_yaml(file)
  
  species_info <- list()
  
  for (entry in data) {
    identifier <- entry$identifier
    species <- mapGenomeBuilds(identifier)
    value <- species[2]
    species_name <- value[1, 1]
    species_name
    chrom <- entry$chrom
    start <- entry$start
    sequence_length <- entry$sequence_length
    strand <- entry$strand
    
    species_info_entry <- list(species_name, chrom, start, sequence_length, strand)
    species_info[[identifier]] <- species_info_entry
  }
  
  return(species_info)
}
```

```r
result <- map_genome_builds("coordinates.yaml")
print(result)
```

## Problem 2 - Extracting DNA Sequences

```r
library(BSgenome.Hsapiens.UCSC.hg38)
library(BSgenome.Mmusculus.UCSC.mm10)
```

```r
library(BSgenome)
library(GenomicRanges)
library(Biostrings)
library(stringr)

extract_dna_sequences <- function(coordinates) {
  sequences <- list()
  
  for (coord_name in names(coordinates)) {
    coord <- coordinates[[coord_name]]
      
    species <- coord[[1]]
    words <- strsplit(species, " ")
    first <- words[[1]][[1]]
    letter <- substr(first, 1, 1)
    second <- words[[1]][[2]]
    processed_species <- paste0(letter, second, sep="")
    
    bs_genome_name <- paste0("BSgenome.", processed_species, ".UCSC.", coord_name)
    genome <- eval(parse(text = bs_genome_name))
    seq <- getSeq(genome, coord[[2]],  coord[[3]],  coord[[3]] + coord[[4]] - 1)
    
    sequences <- c(sequences, list(seq))
  }
  
  return(sequences)
}
```

```r
sequences = extract_dna_sequences(result)
sequences
```

## Problem 3 - Pairwise DNA Sequence Alignment

```r
rownames <- colnames <- c("A", "C", "G", "T")
substitution_matrix <- matrix(
  c(
    3, -1, -1, -1,
    -1, 3, -1, -1,
    -1, -1, 3, -1,
    -1, -1, -1, 3
  ),
  nrow = 4,
  dimnames = list(rownames, colnames)
)

alignment <- pairwiseAlignment(
  pattern = sequences[[1]],
  subject = sequences[[2]],
  substitutionMatrix = substitution_matrix,
  gapOpening = -2
)

alignment
```

The alignment score is 4 which indicates a decent amount of mismatches showing that these sequences are not well aligned. Some notable points of alignment are at the front and the second to last run of matches at the end.

## Problem 4 - Multiple Sequence Alignment and Phylogenetic Tree Construction

```r
library(msa)
```

```r
new_sequences <- readDNAStringSet("multiz30-ace2_partial.fasta")
new_sequences
```

```r
new_sequences <- c(new_sequences, list(sequences[[1]], sequences[[2]]))
new_sequences
```

```r
names(new_sequences)[length(names(new_sequences)) - 1] <- "hg38"
names(new_sequences)[length(names(new_sequences))] <- "mm10"
```

```r
msa_result <- msa(new_sequences)
msa_result
```

```r
library(ape)
```

```r
alignmentMatrix <- msaConvert(msa_result, type = "seqinr::alignment")
alignedDNAbin <- as.DNAbin(alignmentMatrix)
distMatrix <- dist.dna(alignedDNAbin)
njTree <- nj(distMatrix)
library(repr)
options(repr.plot.width=20, repr.plot.height=20, repr.plot.res=200)
plot(njTree, show.tip.label = TRUE)
```
