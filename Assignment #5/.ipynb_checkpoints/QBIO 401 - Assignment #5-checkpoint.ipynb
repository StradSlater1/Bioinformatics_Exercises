# Problem 1 - Getting the Haloptype

```r
gethap <- function(person, side){
    hap <- c()
    for (entry in person){
        hap <- append(hap, c(substr(entry[[1]], (2*side)+1, (2*side)+1)))
    }
    return(hap)
}
```

```r
x22 <- read.csv("abbgen1k.csv", stringsAsFactors=F)
individual <- x22[,10]
gethap(individual, 1)
```

# Problem 2 - Average Pairwise Diversity 

```r
getpi <- function(data, range1, range2, n){
    pair_differences <- numeric(n)
    for (i in seq_len(n)){
        index1 <- sample(range1, 1)
        index2 <- sample(range2, 1)
        if (index1 == index2){
            side = sample(0:1, 1)
            first <- gethap(data[[index1]], side)
            second <- gethap(data[[index2]], side)
        }
        else{
            first <- gethap(data[[sample(range1, 1)]], sample(0:1, 1))
            second <- gethap(data[[sample(range2, 1)]], sample(0:1, 1))
        } 
        differences <- sum(first != second)
        pair_differences[i] <- differences
    }
    return (mean(pair_differences))
}
```

```r
#Did 10 due to the amount of time it took to run >100
getpi(x22, 10:90, 10:90, 10)
```

# Problem 3 - Out-of-Africa Hypothesis

```r
#Did 10 due to the amount of time it took to run >100
print(paste("Euro-Euro: ", getpi(x22, 10:90, 10:90, 10)))
print(paste("Afri-Afri: ", getpi(x22, 91:179, 91:179, 10)))
print(paste("Afri-Euro: ", getpi(x22, 91:179, 10:90, 10)))
```

These values help support the Out-of-Africa hypothesis because accepting the hypothesis would require one to expect less genetic diversity as you extend out of the home land due to things like bottle necks. This is clear from the data due to the fact that there is more genetic diversity between people Africa than people in Europe. Furthermore, the most diversity is between people from Africa and people from Europe due to the fact that they have been geographically isolated from each other and able to change.
