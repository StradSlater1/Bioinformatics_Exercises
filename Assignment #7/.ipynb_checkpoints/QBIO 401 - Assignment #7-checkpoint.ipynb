## Problem 2 - Fasta Files

```r
mad <- read.csv("Mad.txt", header = FALSE, sep = "\t")
max <- read.csv("Max.txt", header = FALSE, sep = "\t")
myc <- read.csv("Myc.txt", header = FALSE, sep = "\t")
```

```r
createFasta <- function(data) {
    file <- ""
    i <- 1
    for (seq in data$V1){
        file <-  paste0(file, paste0(">Sequence", i, "\n"))
        file <- paste0(file, seq, "\n")
        i <- i+1
    }
    return (file)
}
```

```r
writeLines(createFasta(mad), "Mad.fa")
writeLines(createFasta(max), "Max.fa")
writeLines(createFasta(myc), "Myc.fa")
```

## Problem 3 - DNAshapeR 

```r
library(DNAshapeR)
library(Biostrings)
```

```r
madShape <- getShape("Mad.fa")
maxShape <- getShape("Max.fa")
mycShape <- getShape("Myc.fa")
```

## Problem 4 - plotShape() and heatShape() 

```r
options(repr.plot.width=15, repr.plot.height=15, repr.plot.align='center')
par(mfrow = c(2, 2))
plotShape(madShape$MGW, main = "MGW for Mad")
plotShape(madShape$ProT, main = "ProT for Mad") 
plotShape(madShape$Roll, main = "Roll for Mad") 
plotShape(madShape$HelT, main = "HelT for Mad")
par(mfrow = c(1, 1))
options(repr.plot.width=NULL, repr.plot.height=NULL, repr.plot.align=NULL)
```

```r
options(repr.plot.width=15, repr.plot.height=15, repr.plot.align='center')
par(mfrow = c(2, 2))
heatShape(madShape$MGW, 1, main = "MGW for Mad")
heatShape(madShape$ProT, 1, main = "MGW for Mad")
heatShape(madShape$Roll, 1, main = "MGW for Mad")
heatShape(madShape$HelT, 1, main = "MGW for Mad")
par(mfrow = c(1, 1))
options(repr.plot.width=NULL, repr.plot.height=NULL, repr.plot.align=NULL)
```

## Problem 5  - CSV

```r
createCSV <- function(data, name){
    write.csv(data$MGW, paste0(name, "_MGW_.csv"), row.names = FALSE)
    write.csv(data$ProT, paste0(name, "_ProT_.csv"), row.names = FALSE)
    write.csv(data$Roll, paste0(name, "_Roll_.csv"), row.names = FALSE)
    write.csv(data$HelT, paste0(name, "_HelT_.csv"), row.names = FALSE)
}
```

```r
createCSV(madShape, "Mad")
createCSV(maxShape, "Max")
createCSV(mycShape, "Myc")
```
