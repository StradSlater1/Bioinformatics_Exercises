```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
from matplotlib import pyplot as plt
```

```python
dataset = np.load('counts.npy')
labels = np.loadtxt('labels.txt')
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca_results = pca.fit_transform(dataset)
```

```python
import seaborn as sns

df = pd.DataFrame(pca_results, columns = ['PCA 1', 'PCA 2'])
df['Cluster'] = labels

plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x='PCA 1', y='PCA 2', hue='Cluster', palette='viridis', alpha=0.2)

plt.title('PCA visualization of scRNA-seq data')
plt.legend(title='Cluster')
plt.show()
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2)
tsne_results = tsne.fit_transform(dataset)
```

```python
df = pd.DataFrame(tsne_results, columns = ['t-SNE 1', 't-SNE 2'])
df['Cluster'] = labels

plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x='t-SNE 1', y='t-SNE 2', hue='Cluster', palette='viridis', alpha=0.2)

plt.title('t-SNE visualization of scRNA-seq data')
plt.legend(title='Cluster')
plt.show()
```

Both PCA and t-SNE help show that cluster 3 is distinct from cluster 1 and 2. However in the t-SNE graph, the distinction between cluster 1 and 2 is much clearer than in the PCA graph. PCA shows us that 1 and 2 are closer to each other than they each are two 3 while the t-SNE graph helps us see that they are still distinct. t-SNE also shows us that 1 and 2 are more closly packed then the cells in cluster 3.
